```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from Reheat_Assignment import Reheat_Rankine_Cycle
```

```python
# Initialize lists to store reheat pressure and thermal efficiency values
reheat_pressure = []
thermal_efficiency = []

# Loop through pressure values from 0.5 MPa to 12.5 MPa in steps of 0.1 MPa
for i, pressure in enumerate(range(500,12600,100)):
    
    reheat_pressure.append(pressure / 1000) #Convert pressure from MPa to bar
    # Calculate thermal efficiency for the current pressure and append to thermal_efficiency list
    efficiency = Reheat_Rankine_Cycle.Thermal_Efficiency(pressure)
    thermal_efficiency.append(efficiency)


# Create a dictionary with reheat pressure and thermal efficiency
data = {
    'Reheat Pressure (MPa)': reheat_pressure,
    'Thermal Efficiency': thermal_efficiency
}

# Convert the dictionary to a pandas DataFrame
df = pd.DataFrame(data)
print(df)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming your DataFrame is named df and has two columns named 'Column1' and 'Column2'
df.plot(x='Reheat Pressure (MPa)', y='Thermal Efficiency', kind='line')

# Find the maximum point
max_y = df['Thermal Efficiency'].max()
max_x = df.loc[df['Thermal Efficiency'].idxmax(), 'Reheat Pressure (MPa)']

# Plot the maximum point
plt.plot(max_x, max_y, 'ro')

# Annotate the maximum point
plt.annotate(f'({max_x:.2f}, {max_y:.2f})', (max_x, max_y), textcoords="offset points", xytext=(0,-20), ha='center')

# Change x-axis tick locations
plt.xticks(np.arange(0, 12.5, step=1)) 

plt.xlabel('Reheat Pressure (MPa)') # Set lables for x-axis & y-axis 
plt.ylabel('Thermal Efficiency')
plt.show()
```
